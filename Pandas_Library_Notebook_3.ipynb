# **Basics of Time Series**

```python
# Importing pandas for data manipulation and analysis
# Importing numpy for numerical operations and calculations
import pandas as pd
import numpy as np
```

```python
# Reading the Microsoft stock data from a CSV file and storing it in a DataFrame called 'new_df'
new_df = pd.read_csv('/content/Microsoft_Stock.csv')
new_df
```

```python
# Displaying a concise summary of the DataFrame, including:
# - number of non-null entries per column
# - data types of each column
# - memory usage
new_df.info()
```

```python
# Convert 'Date' column to datetime format using the specified format
new_df['Date'] = pd.to_datetime(new_df['Date'], format='%m/%d/%Y %H:%M:%S')
new_df

```

```python
new_df['Date']
```

```python
# Get the day name (e.g., Monday, Tuesday) of the first row's 'Date' value
new_df.loc[0, 'Date'].day_name()
new_df
```

```python
 # Create a new column that stores the day name for each date
new_df['Day_Name'] = new_df['Date'].dt.day_name()
```

```python
 # Get the earliest (minimum) date in the 'Date' column
new_df['Date'].min()
```

```python
# Get the latest (maximum) date in the 'Date' column
new_df['Date'].max()
```

```python
# Calculate the total time span in the dataset
new_df['Date'].max() - new_df['Date'].min()
```

```python
# Filter rows for the year 2020
new_df[(new_df['Date'] >= '2020-01-01') & (new_df['Date'] <= '2020-12-31')]
```

```python
new_df.set_index('Date', inplace=True)
```

```python
# Resample the 'High' column to monthly frequency and get the maximum value in each month
new_df['High'].resample('M').max()
```

```python
# Resampling the 'High' column monthly and plotting the maximum high for each month
new_df['High'].resample('M').max().plot()
```

```python
# Resample the data to monthly frequency and calculate:
# - average 'Open' and 'Close' prices,
# - maximum 'High' price,
# - minimum 'Low' price,
# - total 'Volume' traded each month
new_df.resample('M').agg({'Open': 'mean', 'High': 'max', 'Low': 'min', 'Volume': 'sum', 'Close': 'mean'})
```
